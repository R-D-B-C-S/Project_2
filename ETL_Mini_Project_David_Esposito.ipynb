```python
temp_container = crowdfunding_info_df[['category & sub-category']]
temp_container

new_columns = []
for col in temp_container:
    new_columns.append(col.split("&"))
print(new_columns)
```

```python
new_rows= []
for row in temp_container.values.tolist():
    new_rows.append(row[0].split('/'))
#print(new_rows)

category_subcategorydf = pd.DataFrame(new_rows, columns = new_columns)
category_subcategorydf
```

```python
crowdfunding_info_df_categories = crowdfunding_info_df.join(category_subcategorydf)


crowdfunding_info_df_categories.head()
```

```python
crowdfunding_info_df_categories.rename(columns = {'(category ,)':'category'}, inplace = True)

crowdfunding_info_df_column_names = list(crowdfunding_info_df.columns)
category_subcategorydf_column_names = list(category_subcategorydf.columns)

crowd_with_categories = crowdfunding_info_df.join(category_subcategorydf)
crowd_with_categories.columns = crowdfunding_info_df_column_names + ['category', 'sub-category']


crowd_with_categories.head()
```

```python
# Assign the category and subcategory values to category and subcategory columns.
```

```python
#crowd_with_categories['category'].value_counts(dropna=False)
categories_list=[]
for element in crowd_with_categories['category']:
    if element not in categories_list:
        categories_list.append(element)
print("categories ",categories_list)
subcategory_list=[]
for subelement in crowd_with_categories['sub-category']:
    if subelement not in subcategory_list:
        subcategory_list.append(subelement)
print("sub categories ",subcategory_list)

```

```python
# Get the unique categories and subcategories in separate lists.

```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories_list))
print(len(subcategory_list))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
cat_ids = [str(x)+"cat" for x in category_ids]


scat_ids=[str(x)+"subcat" for x in subcategory_ids]
print(cat_ids)
print(scat_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id. 
# Use a list comprehension to add "subcat" to each subcategory_id.    

    
print(cat_ids)
print(scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.


category_dict ={'category_id':cat_ids,
               'category':categories_list}

category_df_complete=pd.DataFrame(category_dict)
category_df_complete.head()
# Create a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
subcategory_dict ={'subcategory_id':scat_ids,
               'sub-category':subcategory_list}

subcategory_df_complete=pd.DataFrame(subcategory_dict)
subcategory_df_complete.head()
```
